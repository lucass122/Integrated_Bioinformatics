## Convert Data from Pandas to d3.js
To work with d3.js, it's best to format data tables (such those obtained from Pandas) into arrays of Objects.
This is how a JSON example would look like:
```javascript
[
    {column1: value, column2: value, ...}  // row 1
    {column1: value, column2: value, ...}, // row 2
    {column1: value, column2: value, ...}, // row 3
]
```

### Write a function that converts a Pandas dataframe to an array of objects as described above.
It's up to you if you want to use Python or JavaScript for that.
Note that the [to_json](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_json.html) is useful to get your data into JavaScript, but see yourself whether this is what you want or not...

````python
%%javascript
require(['d3'], function(d3){
function convertPlotData(data){
  var convertedData = [];
  //pandas gives each column the same number of
  //objects, arbitrarily choose one to iterate over
  for(var i in data[Object.keys(data)[0]]){
    var convertedDatum = {};
    for(var key in data){
      convertedDatum[key] = data[key][i];
    }
    convertedData.push(convertedDatum);
  }
  return convertedData;
}

window.convertPlotData = convertPlotData;
});
````

````python
%%javascript
require(['d3'], function(d3){

  //a weird idempotency thing
  $("#chart3").remove();
  //create canvas
  element.append("<div id='chart3'></div>");
  $("#chart3").width("820");
  $("#chart3").height("620");        
  var margin = {top: 20, right: 20, bottom: 30, left: 40};
  var width = 800 - margin.left - margin.right;
  var height = 600 - margin.top - margin.bottom;
  var svg = d3.select("#chart3").append("svg")
    .style("position", "relative")
    .style("max-width", "800")
    .attr("width", width + "px")
    .attr("height", (height + 50) + "px")
    .append("g")
    .attr("transform", "translate(" + margin.left + "," + margin.top + ")");
  
  //set data
  var data = window.convertPlotData(window.volcano);
  var xVals = function(d){return d.log2_fold_change};
  var yVals = function(d){return d["-log_p_values"]};
  
  var xScale = d3.scale.linear().range([0, width]);
  var xAxis = d3.svg.axis().scale(xScale).orient("bottom");
  var xMap = function(d) { return xScale(xVals(d));};
      
  var yScale = d3.scale.linear().range([height, 0]);
  var yAxis = d3.svg.axis().scale(yScale).orient("left");
  var yMap = function(d) { return yScale(yVals(d));};
  xScale.domain([d3.min(data, xVals)-1, d3.max(data, xVals)+1]);
  yScale.domain([d3.min(data, yVals)-1, d3.max(data, yVals)+1]);

  
    var cValue = function(d) { 
    if(d["-log_p_value"] < 0 ){
        return "red";
    }
        else{
    return "orange";}
  
  }
  
  // x-axis
  svg.append("g")
      .attr("class", "x axis")
      .attr("transform", "translate(0," + height + ")")
      .call(xAxis)
    .append("text")
      .attr("class", "label")
      .attr("x", width - 80)
      .attr("y", -15)
      .style("text-anchor", "end")
      .text("log2 fold change");

  // y-axis
  svg.append("g")
      .attr("class", "y axis")
      .call(yAxis)
    .append("text")
      .attr("class", "label")
      .attr("transform", "rotate(-90)")
      .attr("y", 0)
      .attr("dy", "1em")
      .style("text-anchor", "end")
      .text("-log10 p value");
    
    var leftLine = d3.select('body').append("line")
    
    
    
  //NEW: TOOLTIP. 
  var tooltip = d3.select("body").append("div")
    .attr("class", "tooltip")
    .style("opacity", 0)
    .style("background-color", "white");
 
  svg.selectAll(".dot")
     .data(data)
     .enter().append("circle")
     .attr("class", "dot")
     .attr("r", 2.5)
     .attr("cx", xMap)
     .attr("cy", yMap)
     .style("fill", "orange") //D3 does the magic! 
     .on("mouseover", function(d) { //much like jquery, an event listener
         tooltip.transition()
                .duration(50)
                .style("opacity", .9);
         tooltip.html(d["Gene"])
               .style("left", (d3.event.pageX + 5) + "px")
               .style("top", (d3.event.pageY - 28) + "px");
      })
      .on("mouseout", function(d) {
          tooltip.transition()
               .duration(500)
               .style("opacity", 0);
      });
    
 
});
````
