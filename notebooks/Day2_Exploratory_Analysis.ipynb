### Step 1: convert cell values to numbers

By replacing the letters B, D, H and U with the corresponding numbers 1, 2, 3 and 4 the data can be further processed.

```python
for col in bxdGenotype:
    if col.startswith('BXD'):
      bxdGenotype[col] = bxdGenotype[col].replace({'B':1 ,'D':2,'H':3,'U':4})
    bxdGenotype[1:10]
```

### Subtask: Compare the two different diets (for several biochemistry measurements)
Tip: after prototyping your analysis, wrap your individual steps in a function that you can use for an arbitrary phenotype in an arbitrary worksheet

Mean glucose blood values are compared between the two diets first. AS the bar plots indicates they don't differ significantly.

```python
hfd_glucose = phenotype['Biochemistry']['HFD_Glucose_[mmol/L]'].mean()
cd_glucose =  phenotype['Biochemistry']['CD_Glucose_[mmol/L]'].mean()
test = pd.DataFrame(data=[cd_glucose,hfd_glucose],index=['CD Glucose','HFD Glucose'])
hfd_glucose,cd_glucose
```

```python
test.plot.bar()
```

The cholesterol value is compared next an as expected the high fat diet corresponds to a higher mean cholesterol value.

```python
hfd_cholesterol = phenotype['Biochemistry']['HFD_Cholesterol_[mmol/L]'].mean()
cd_cholesterol =  phenotype['Biochemistry']['CD_Cholesterol_[mmol/L]'].mean()
test = pd.DataFrame(data=[cd_cholesterol,hfd_cholesterol],index=['CD Cholesterol','HFD Cholesterol'])
hfd_cholesterol,cd_cholesterol
```

```python
test.plot.bar()
```

__Define functions that perform aspects of the procedure above__

```python
def compareDiet(diet1,diet2):
     diet1_mean = phenotype['Biochemistry'][diet1].mean()
     diet2_mean = phenotype['Biochemistry'][diet2].mean()
     test = pd.DataFrame(data=[diet1_mean,diet2_mean],index=[diet1,diet2])
     test.plot.bar()
     return (diet1_mean,diet2_mean)
    
    
```

__After refactoring code into functions: Apply function to other phenotypes__

The blood protein value doesn't show any significant difference as well between the two diets.

```python
compareDiet('CD_BloodProtein_[g/L]','HFD_BloodProtein_[g/L]')
```

The creatine kinase value, however is higher for the mice with the common diet.

```python
compareDiet('CD_CreatineKinase_[U/L]','HFD_CreatineKinase_[U/L]')
```

# Plot the linear dependence of pairs of phenotypes


To compare the linear dependence of two data sets a regression plot is drawn. In this case this is done with the CD diet iron values and hemoglobin to see if there is a correlation. As the plot indicates some values are located on or near the regression line, however not all the values can be found in this location, so no real conclusion can be drawn from the plot.

```python
cdiron = phenotype['Biochemistry']['CD_Iron_[umol/L]']
cdhemo = phenotype['Hematology']['CD_HGB.Hemoglobin_[g/dL]']
test = pd.DataFrame(data=[cdiron,cdhemo])
```

```python
sns.regplot(x=cdiron, y=cdhemo, data=test)
```
